```python
import graphviz

g = graphviz.Digraph()
g.attr('node', shape='circle')
transition_matrix = [[0,1,0,0,3,0,0],
                     [0,0,3,2,1,0,0]]

for rows in range(len(transition_matrix)):
    for columns in range(len(transition_matrix[rows])):
        if transition_matrix[rows][columns] !=0:
            g.edge(str(rows),str(columns),label=str(transition_matrix[rows][columns]))
        



g
```
